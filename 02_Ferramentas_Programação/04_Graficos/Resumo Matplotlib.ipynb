## API FUNCIONAL

```python
x = [1, 2, 3, 4, 5]
y1 = [n ** 2 for n in x]
y2 = [n ** 3 for n in x]
y3 = [n ** 4 for n in x]
```

### No mesmo eixo

```python
plt.plot(x, y1, color='red', linestyle='--', linewidth=5, label='f(x) = x^2')
#label = rótulo
#Nesse caso é o rótulo desse gráfico
plt.plot(x, y2, color='#27a16c', marker='o', label='f(x) = x^3')
#Hexadecimal: https://htmlcolorcodes.com/

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Comparação entre cubo e quadrado')
plt.grid(alpha=0.3)
plt.legend(); #Faz aparecer os rótulos dos gráficos
#Em uma mesma linha:
#plt.legend(['f(x) = x^2', 'f(x) = x^3'], ncol=2, loc='upper left')
```

```python
plt.style.use('classic')
#JEITO DE PASSAR TAMANHO P/ FIGURE NA API FUNCIONAL
plt.figure(figsize=(5,6))
plt.plot(x, y1)
#COMO MUDAR ESCALA DO EIXO
plt.yticks([i for i in range (0, 26, 2)]);
```

## API ORIENTADA A OBJETOS

### No mesmo eixo (mesmo gráfico)

```python
fig, ax = plt.subplots(figsize=(10, 6)) #Largura, altura

ax.plot(x, y1, color='red', linestyle='--', linewidth=5, label='f(x) = x^2')
ax.plot(x, y2, color='#27a16c', marker='o', label='f(x) = x^3')

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Comparação entre cubo e quadrado')
#Em uma mesma linha:
#ax.set(xlabel='x', ylabel='f(x)', title='Comparação entre cubo e quadrado')
ax.legend();
#Em uma mesma linha:
#ax.legend(['f(x) = x^2', 'f(x) = x^3'], ncol=2, loc='upper left')
```

### Múltiplos eixos

```python
figure = plt.figure(figsize=(10,6))
ax1 = figure.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = figure.add_axes([0.2, 0.5, 0.3, 0.3])
#largura e altura p/ definir o tamanho do retângulo
#Esquerda e inferior p/ saber onde posicionar ele
#Porcentagens dos comprimentos da figure

figure.patch.set_facecolor('blue') #Mudar a cor da figure
ax2.set_facecolor('xkcd:salmon') #Muda a cor dos axes

ax1.plot(x, y1, color='red', ls='--', lw=5, label='f(x) = x^2')
ax2.plot(x, y2, color='#27a16c', marker='o', label='f(x) = x^3');
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, sharey=True) #ou simplesmente(1, 2)
fig.suptitle('1 X 2 SUBPLOTS', fontsize=30)
fig.tight_layout(pad=2);
#pad: Preenchimento entre a borda da figura e as bordas dos subtramas

axs[0].text(1.9, 65, 'gráfico[0]', fontsize=16)
axs[1].text(1.5, 65, 'gráfico[1]', fontsize=16) #Coordenadas dos axes p/ começar o texto

axs[0].plot(x, y1)
axs[1].plot(x, y2);
```
