### **O que são os sistemas de controle de versão?**

São ferramentas que funcionam ao mesmo tempo como máquinas do tempo e robôs de integração para o nosso código. Isto porque nos permitem ter acesso a versões antigas do código e também a integração entre alterações efetuadas no código por diferentes pessoas programadoras.

Nos sistemas de controle de versão, há um repositório que nos permite obter qualquer versão já existente do código. Sempre que quisermos controlar as versões de algum arquivo, temos que informar que queremos rastreá-lo no repositório. A cada mudança que desejamos efetivar, devemos armazenar as alterações nesse repositório.

Alterações nos mesmos arquivos são mescladas de maneira automática sempre que possível. Já possíveis conflitos são identficados a cada vez que obtemos as mudanças dos nossos colegas de time.



### **O que é o Git?**

O Git é um sistema de controle de versão (repositório local) que, pela sua estrutura interna, é uma máquina do tempo extremamente rápida e é um robô de integração bem competente.

### **O que é o GitHub?**

Criado em 2008, o GitHub é uma aplicação Web (repositório remoto) que possibilita a hospedagem de repositórios Git, além de servir como uma rede social para programadores.

### **Como versionar um arquivo e enviá-lo para um repositório no GitHub?**

1) Certifique-se de estar na pasta correta, com os arquivos que deseja enviar para o GitHub. Nesta pasta, abra o Git Bash (ou use o prompot de comando).

2) Vamos transformar a pasta atual em um repositório Git, então execute a linha de comando:
```
git init
```
Uma pasta oculta git foi criada e a pasta já se tornou um repositório local.

3) Para entender a situação dos arquivos e se eles já começaram a ser versionados, utilize o comando:
```
git status
```
4) Para que um arquivo comece a ser rastreado pelo Git, devemos ir pela seguinte linha:
```
git add . | (para toda a pasta)
git add nome do arquivo | (quando se trata de apenas um arquivo)
```
5) Se dermos um novo `git status` veremos que o arquivo ou a pasta estão sendo rastreados pelo Git, mas ainda não foram gravados (ou comitados, em linguagem mais técnica) no repositório. Para isto, utilize o comando:
```
git commit -m "comentário sobre o primeiro commit"
```
Esse `m` serve para informar a mensagem do commit.

6) Caso alguma nova alteração na pasta ou em algum dos arquivos seja feita, devemos seguir com os dois comandos:
```
git add nome do arquivo ou do diretório
git commit -m "nova alteração"
```
7) Para verificar o histórico das alterações gravadas no repositório, podemos executar o comando:
```
git log
```
8) Com as alterações feitas, é hora de apontarmos o repositório local para o repositório remoto do GitHub:
```
git remote add origin link_https_ou_ssh_referente_ao_repositorio
git push origin main
```

### **Como clonar um repositório do GitHub?**

1) Crie uma pasta na sua máquina e entre nela, seja pelo bash ou pelo terminal.

2) Execute o comando:
```
git clone link_https_ou_ssh_referente_ao_repositorio
```
3) O repositório se transforma em uma subpasta da pasta que temos na nossa máquina. Se executarmos um `git log` nesta subpasta, será possível ter acesso a todo o histórico de commits.

### **Algo interessante de ser dito...**

Quando informamos para o Git que queremos rastrear um arquivo, executando `git add` pela primeira vez, o Git coloca esse arquivo em uma área especial do repositório, chamada de *stage*.

Uma vez que um arquivo está na área de stage, todas as mudanças nesse arquivo passam a ser examinadas.

O diretório que contém nosso arquivos é chamado de diretório de trabalho, ou *working directory*.

### **Como criar uma branch?**

A maioria dos sistemas de controle de versão permite trabalho paralelo por meio de *branches*. Uma *branch* é uma linha independente de desenvolvimento em que podemos comitar novas versões do código sem afetar sem afetar outras *branches*.

O Git possui por padrão uma branch principal chamada *main*. 

Para criar uma nova *branch* no local, o seguinte deve ser feito:
```
git checkout -b nome_da_sua_branch
```
Já no remoto:
```
git push --set-upstream origin nome_da_sua_branch
```
Para trocar de branch, basta utilizar:
```
git checkout nome_da_branch
```
Para checar em qual branch você está:
```
git branch -a
```


### **Atualizei um arquivo no local e quero passar isso para o remoto, e agora?**

Primeiro, passe suas atualizações para o remoto (dentro da sua *branch*):
```
git add .
git commit -m "sua alteração"
git push
```

Depois, vamos atualizar o repositório local com o que está na main remota:
```
git checkout main
git pull
```

Com estas atualizações feitas, falta um update na sua branch remota com o que está na main, para isso vamos fazer um merge entre as duas:
```
git checkout nome_da_sua_branch
git merge main
```

Se não houver conflito, é hora de empurrar esse repositório local que mescla minhas mudanças e mudanças na main para a minha branch remota:
```
git push
```
