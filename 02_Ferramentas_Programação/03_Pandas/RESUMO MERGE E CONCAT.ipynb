# MERGE

Ele usa as colunas iguais p/ ser a "chave", a não ser que vc especifique quais colunas vc quer

```python
df1 = pd.DataFrame({"RA": [1, 4, 2, 3, 5, 6, 6],
                   "Aluno": ["joão", "leia", "maria", "han", "luke", "anakin", "carol"],
                   "Prova 1": [10, 10, 9, 8, 7, 10, 8]})
df1
```

```python
df2 = pd.DataFrame({"RA": [1, 4, 2, 3, 5, 7, 6],
                   "Aluno": ["joão", "leia", "maria", "han", "luke", "obi wan", "ju"],
                   "Prova 2": [10, 10, 9, 8, 7, 10, 8]})
df2
```

```python
df1.merge(df2, how="outer")
```

```python
df1.merge(df2, how="outer", on="RA", suffixes=['_prova1', '_prova2']) #Ele trata como se Aluno fossem colunas distintas nos dfs
```

```python
df1.merge(df2, how="inner")
```

```python
df1.merge(df2, how="left")
```

```python
df1.merge(df2, how="right")
```

```python
df2.rename({'RA':'RA_p2', 'Aluno':'aluno_p2'}, axis=1, inplace=True)
df2
```

```python
df1.merge(df2, how='outer', left_on=['RA','Aluno'], right_on=['RA_p2', 'aluno_p2'], indicator=True)
#TRATA COMO SE AS COLUNAS TIVESSEM OS MESMOS RESPECTIVOS NOMES, MAS COLOCA DUAS COLUNAS
```

# CONCAT

```python
data1 = pd.DataFrame({"col1":["1A", "1B", "1C", "1D", "1E", "1F"],
                     "col2":["2A", "2B", "2C", "2D", "2E", "2F"]})
data1
```

```python
data2 = pd.DataFrame({"col1":["1A", "1B", "1C", "1D"],
                     "col3":["2A", "2B", "2C", "2G"]},
                    index=[4,5,6,7])
data2
```

```python
pd.concat([data1, data2], join="outer") #Padrão, pega todas as colunas
```

```python
pd.concat([data1, data2], join="inner") #Pega só as colunas pertencentes aos dois
```

```python
pd.concat([data1, data2], ignore_index=True) #Faz a mesma coisa só que renomeia os índices p padrão
```

```python
pd.concat([data1, data2], verify_integrity=True)
# Dá erro pq tem índices duplicados (4 e 5)
```

```python
pd.concat([data1, data2], keys=["Data 1", "Data 2"], names=["df", "índices"], verify_integrity=True) #keys explicitam da onde vem
#names dão nome aos índices
```
