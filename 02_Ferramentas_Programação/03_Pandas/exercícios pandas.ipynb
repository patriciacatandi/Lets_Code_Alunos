```python
tabela = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None, names=["sepal_length", "sepal_width", "petal_length", "petal_width", "class"])
tabela
```

c. Calcule a média de cada coluna numérica para cada um dos tipos de íris indicados na coluna "class" sem utilizar métodos que não foram ensinados na aula de hoje. Utilize o loop for para reduzir quantidade de linhas. <br>
Existe diferença entre elas?

```python
for classe in ["Iris-setosa", "Iris-virginica", "Iris-versicolor"]:
    print(classe)
    print(tabela[tabela["class"] == classe].mean()) #Não dá erro em colunas não numéricas, ele simplesmente não faz
    print()
```

d. Adicione uma única coluna com a média de 'sepal length' para cada um dos tipos de íris indicados na coluna "class". Utilize o loop for e faça isso sem utilizar métodos que não foram ensinados na aula de hoje.

```python
for classe in ["Iris-setosa", "Iris-virginica", "Iris-versicolor"]:
    tabela.loc[tabela["class"] == classe, "media_sepal_legth"] = tabela[tabela["class"] == classe]["sepal_length"].mean()

tabela
```

e. Crie uma coluna de volume sabendo que o volume representa (pi x petallength x sepal_length^2)/3

```python
import numpy as np

tabela["volume"] = (np.pi * tabela["petal_length"] * (tabela["sepal_length"] ** 2)) /3

tabela
```
