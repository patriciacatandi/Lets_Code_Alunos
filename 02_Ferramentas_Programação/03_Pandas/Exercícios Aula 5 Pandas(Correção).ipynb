1. Baixe os dados de consumo de bebidas por país do kaggle faça uma análise das informações utilizando os métodos que você já conhece e depois responda:

```python
drinks = pd.read_csv(r"C:\Users\Carol\Desktop\Lets_Code_Alunos\02_Ferramentas_Programação\03_Pandas\data\drinks.csv")
drinks
```

a. Encontre qual a bebida mais consumida em cada um dos países e a quantidade.



```python
# Seleciona a bebida mais consumida
df_bebida = (drinks
             .set_index(['country','continent'])
             .idxmax(axis=1) #Pega o índice (nome da coluna) do max por linha
             .rename('bebida', axis=1) #Como só tem uma coluna, pode passar o rename com o novo nome direto
             .reset_index())
df_bebida
```

```python
#Seleciona a quantidade da bebida mais consumida
df_quantidade = (drinks
                 .set_index(['country','continent'])
                 .max(axis=1) #Pega o valor max por linha
                 .rename('quantidade', axis=1)
                 .reset_index())
df_quantidade
```

```python
# Junta os dois df
df_bebida.merge(df_quantidade, how='inner') #Juntou usando as duas colunas iguais do df como "chave"
#Olhou pras chaves que tem nos dois (que por coincidência são todas)
#Nesse caso poderia ser qualquer how
```

```python
# com os próximos exercícios faremos esse mesmo processo de forma mais fácil
```

b. Crie um df cujas bebidas estejam agrupadas em uma mesma coluna.

```python
drinks_melt = drinks.melt(id_vars=['country','continent'], value_name='volume', var_name='drink')
drinks_melt
```

c. Utilizando esse novo df, encontre qual a bebida mais consumida por país e a quantidade.

```python
indices = drinks_melt.groupby("country").idxmax().volume
indices
```

```python
drinks_melt.loc[indices]
```
